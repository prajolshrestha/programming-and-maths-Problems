## 2D Transformation matrices

### 1.Rotation and Stretching

```python
#2d input vector
v = np.array([3, -2])

# 2 x 2 transformation matrix
A = np.array([ [1,-1], [2,1] ])

# output vector is Av
w = A@np.matrix.transpose(v)

# plot
plt.plot([0,v[0]],[0,v[1]],label='v')
plt.plot([0,w[0]],[0,w[1]],label='Av')
plt.grid()
plt.title('Rotation and Stretching')
plt.axis((-6,6,-6,6))
plt.legend()
plt.show()
```

### 2.Pure Rotation

```python
#input vector
v = np.array([2,0])

#rotation matrix
theta = np.pi/2
A = np.array([[math.cos(theta),-math.sin(theta)],
              [math.sin(theta),math.cos(theta)]])

#output
w = A@np.matrix.transpose(v)

#plot
plt.plot([0,v[0]],[0,v[1]],label='v')
plt.plot([0,w[0]],[0,w[1]],label='Av')
plt.grid()
plt.title('Rotation ')
plt.axis((-6,6,-6,6))
plt.legend()
plt.show()
```

### code challange: pure vs impure rotation

```python
v = np.array([2,0])

thetas = np.linspace(0,2*np.pi,100)

vec_mag = np.zeros((len(thetas),2))

for i in range(len(thetas)):
  th = thetas[i]
  A1 = np.array([[2*math.cos(th),-math.sin(th)],
                [math.sin(th),math.cos(th)]])
  A2 = np.array([[math.cos(th),-math.sin(th)],
                [math.sin(th),math.cos(th)]])
  
  #vector magnitude
  vec_mag[i,0] = np.linalg.norm(A1@v.T)
  vec_mag[i,1] = np.linalg.norm(A2@v.T)
  

plt.plot(thetas,vec_mag,'o-')
plt.xlabel('Rotation angle')
plt.ylabel('Av Magnitude')
plt.legend(['Impure Rotation','pure Rotation'])
plt.show()
```

Conclusion: Impure rotation creates Rotation and Stretching

### code challange: Fourier transform via matrix multiplication

```python
%matplotlib inline
#Create fourier matrix
n = 52
w = np.exp((-2*np.pi*1j)/n)
F = np.zeros((n,n), dtype='complex')

for j in range(n):
  for k in range(n):
    m = j*k #j and k starts from 0
    F[j,k] = pow(w,m)

#random vector
x = np.random.randn(n)
print("F:",F)
print("x:",x)

#Mat mul Fourier transform
x1 = F @ x
print(x1)
#FFT
x2 = np.fft.fft(x)
print(x2)

# plot them
plt.figure()
plt.plot(abs(x1))
plt.plot(abs(x2), marker='o', color='r', linestyle='None')#FFT
plt.show()

fig, ax = plt.subplots()
im = ax.imshow(np.real(F))

fig, ax = plt.subplots()
im = ax.imshow(np.imag(F))

fig, ax = plt.subplots()
im = ax.imshow(abs(F))

#plt.show()
```

### code challenge: Matric asymmetry index

```python
## part1: implement MAI

def MAI(A):
  Aanti = (A-A.T)/2
  mai = np.linalg.norm(Aanti) / np.linalg.norm(A)
  return mai
```

```python
## part 2: compute MAI for symmetric, skew.symmetric and random matrix

A = np.random.randn(5,5)
A = (A+A.T)/2 #symmetric matrix
print("Symmetric matrix norm ratio:",MAI(A))

A = np.random.randn(5,5)
A = (A-A.T)/2 #Asymmetric matrix
print("Asymmetric matrix norm ratio:",MAI(A))

A = np.random.randn(5,5) #Random matrix
print("Random matrix norm ratio:",MAI(A))
```
