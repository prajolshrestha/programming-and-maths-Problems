## Import

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
from matplotlib.figure import projections
# Matrix which defines a plane
S = np.vstack([[3,5,1],[0,2,2]]).T

#vector
v = np.array([[-3],[1],[5]]) # in the column space
#
v = np.array([[1],[7],[3]]) # not in the column space


fig = plt.figure()
ax = plt.axes(projection= '3d')

#draw plane according to columns
xx, yy = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10)) # grid banako
cp = np.cross(S[:,0],S[:,1])  # cross product
z1 = (-cp[0]*xx - cp[1]*yy)/cp[2] #normal vector of plane
ax.plot_surface(xx,yy,z1,alpha=.2)

#plot two vectors
plt.plot([0,S[0,0]], [0,S[1,0]], [0, S[2,0]], 'k')
plt.plot([0,S[0,1]], [0,S[1,1]], [0, S[2,1]], 'k')

# plot vector
plt.plot([0,v[0]],[0,v[1]],[0,v[2]], 'r')

ax.view_init(elev=150,azim=0) # change elev for better view
plt.show()

#conclusion: vector is in the span of columns of matrix
```
