```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
# singular matrix(reduced rank matrix) has determinant of 0

# generate 2x2 matrix of integers and with linear dependencies
A = np.random.randint(1,5,(2,2))
A[:,-1] = 2*A[:,-2]

rank = np.linalg.matrix_rank(A)
print(rank)
det = np.linalg.det(A)
print(det)

# m*n matrix
m = 300
A = np.random.randint(1,5,(m,m))
A[:,-1] = 2*A[:,-2]

rank = np.linalg.matrix_rank(A)
print(rank)
det = np.linalg.det(A)
print(det)
```

```python
## det with row swap

A = np.random.randn(6,6)
print('Before row swap:' +str(np.linalg.det(A)))

A1 = A[[1,0,2,3,4,5],:]#first and second row swapped
print('After 1 row swap:' +str(np.linalg.det(A1)))

A2 = A[[1,0,2,4,3,5],:]#first, second and 3rd and 4th row swapped
print('After 2 row swap:' +str(np.linalg.det(A2)))
```

```python
# Determinant of shifted matrix
m = 20

l = np.linspace(0,.1,30)

#initialize
tmp = np.zeros(1000)
dets = np.zeros(len(l))

for deti in range(len(l)):
  
  #run 1000 iteration
  for i in range(1000):
    M = np.random.randn(m,m)
    M[:,0] = M[:,1] 

    #compute determinant
    tmp[i] = abs(np.linalg.det(M + l[deti] * np.eye(m))) # adding identity matrix
  
  #compute average determinant
  dets[deti] = np.mean(tmp)

plt.plot(l,dets,'s-')
plt.xlabel('Fraction of identity for shifting')
plt.ylabel('Determinant')
plt.legend()
plt.show()

# conclusion: The more you add identity matrix, higher determinant you get
```

## code challenge: determinant of matrix product

```python
# det(AB) = det(A)*det(B)

# for random 3x3 matrices
A = np.random.randn(3,3)
B = np.random.randn(3,3)
AB = A@B

print(np.linalg.det(A),np.linalg.det(B),
      np.linalg.det(A)*np.linalg.det(B),
      np.linalg.det(AB))

#loop random matrix size up to 40x40
dets = np.zeros((40,2))

for k in range(40):
  A = np.random.randn(k,k)
  B = np.random.randn(k,k)
  AB = A@B

  dets[k,0] = np.linalg.det(A)*np.linalg.det(B)
  dets[k,1] = np.linalg.det(AB)

plt.plot(dets[:,0]-dets[:,1],'s-')
plt.ylim([-1,1])
plt.show()
  
#conclusion: Theroiticaly we expect the plot of the difference to be a straight line at 0.
          #But, It fails in ral life after matrix size 10--15
```
