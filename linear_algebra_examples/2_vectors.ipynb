## Dot product properties

### 1. Distributive property

```python
# create random vectors
n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

# a(b+c) = a.b + a.c
res1 = np.dot(a,(b+c))
res2 = np.dot(a,b)+np.dot(a,c)
res1, res2 #equal
```

### Associative property

```python
# create random vectors
n = 10
a = np.random.randn(n)
b = np.random.randn(n)
c = np.random.randn(n)

res1 = np.dot(a, np.dot(b,c))
res2 = np.dot(np.dot(a,b),c)

res1, res2 #not equal
```

### commutive property

```python
a = np.random.randn(10)
b = np.random.randn(10)

res1 = np.dot(a,b)
res2 = np.dot(b,a)
res1, res2# equal
```

## dot product from geometric perspective

```python
from matplotlib import projections
v1 = np.array([16,-2,4])
v2 = np.array([.5,2,-1])

#compute angle (radians)
ang = np.arccos(np.dot(v1,v2) / (np.linalg.norm(v1)*np.linalg.norm(v2)))

# plot
fig = plt.figure(figsize=plt.figaspect(1))
ax = plt.axes(None, projection = '3d')
ax.plot([0,v1[0]],[0,v1[1]], [0,v1[2]],'b')
ax.plot([0,v2[0]],[0,v2[1]],[0,v2[2]],'r')

plt.axis((-6,6,-6,6))
plt.title('Angle between vectors: %s rad.' %ang)
plt.show()
```

```python
#equvalence of algebric and geometric dot product formulas

dp_a = np.dot(v1, v2)

dp_g = np.linalg.norm(v1)*np.linalg.norm(v2)*np.cos(ang)

dp_a, dp_g
```

## cauchy-schwarz inequality

```python
# |A'b| <= ||a|| ||b||

#create 3 vectors, one of which is dependent on another
a = np.random.randn(5)
b = np.random.randn(5)
c = np.random.randn(1) * a

# compute dot product
aTb = np.dot(a,b)
aTc = np.dot(a,c)

#demponstrate (in)equalities
print(f'{np.abs(aTb):.4f}, {np.linalg.norm(a)*np.linalg.norm(b):.4f}') #inequality
print(f'{np.abs(aTc):.4f}, {np.linalg.norm(a)*np.linalg.norm(c):.4f}') #equality

```

## Outer Product

```python
v1 = np.random.randn(6)
v2 = np.random.randn(5)

op1 = np.outer(v1,v2) #method 1

#method 2
op2 = np.zeros((len(v1),len(v2)))
for i in range(0,len(v1)):
  for j in range(0,len(v2)):
    op2[i,j] = v1[i] * v2[j]

op1, op2
```

## Cross Product

```python
v1  = [ -3,  2, 5 ]
v2  = [  4, -3, 0 ]

cp1 = np.cross(v1, v2) #method 1

#method 2
cp2 = [[v1[1]*v2[2] - v1[2]*v2[1]],
       [v1[2]*v2[0] - v1[0]*v2[2]],
       [v1[0]*v2[1] - v1[1]*v2[0]]]

cp1, cp2
```

## Hermitian Transpose( Conjugate Transpose)

```python
#create complex number
z = complex(3,4)
print(z)

# magnitude
n = np.linalg.norm(z)
print(n)

# transpose
zT = np.transpose(z)*z
print(zT)

# Hermition Transpose
h = np.transpose(z.conjugate())*z
print(h)




# complex vector
cv = np.array([3, 4j, 5+2j, complex(2,-5)])

#transpose
print(cv.T)#method 1
print(np.transpose(cv))#method 2

# hermition transpose
print(np.transpose(cv.conjugate()))

cv_distance = (np.transpose(cv.conjugate()))*cv
print(cv_distance)
```

## Span

```python
# set S
S1 = np.array([1, 1, 0])
S2 = np.array([1, 7, 0])

# vectors v and w
v = np.array([1, 2, 0])
w = np.array([3, 2, 1])

print("s1: %s , S2: %s , v: %s , w: %s" % (S1,S2,v,w))

#plot
ax = plt.axes(projection='3d')
ax.plot([0,S1[0]],[0,S1[1]],[.1, S1[2]+.1],'r',linewidth=3)# .1 is just to alivate from plane to visualize better
ax.plot([0, S2[0]],[0, S2[1]],[.1, S2[2]+.1],'r',linewidth=3)

ax.plot([0, v[0]],[0, v[1]],[.1, v[2]+.1],'g',linewidth=3) # it is in the span
ax.plot([0, w[0]],[0, w[1]],[0, w[2]],'b') # it is not in the span

#plane
xx, yy = np.meshgrid(range(-15,16), range(-15,16))
cp = np.cross(S1,S2)
z1 = (-cp[0]*xx - cp[1]*yy)*1./cp[2]
ax.plot_surface(xx,yy,z1)

plt.show()
```
